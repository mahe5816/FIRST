```python
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
data=iris.data
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
s=SVC(kernel='linear',C=30)
s.fit(x_train,y_train)
s.score(x_test,y_test)
```

```python
cross_val_score(SVC(kernel='linear',C=30),iris.data,iris.target,cv=5)
```

```python
import numpy as np
ker=['linear','rbf']
c=[1,10,20]
avg={}
for i in ker:
    for j in c:
        a=cross_val_score(SVC(kernel=i,C=j),iris.data,iris.target,cv=5)
        avg[i+'_'+str(j)]=np.average(a)
avg
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
cv=GridSearchCV(SVC(),{
    'C':[1,10,20],
    'kernel':['linear','rbf']
},cv=5)
cv.fit(iris.data,iris.target)
cv.cv_results_
```

```python
dd=pd.DataFrame(cv.cv_results_)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
model_param= {
    'Logistic_Regression': {
        'model':LogisticRegression(),
        'params':{
        'C':[1,10,20]
          }
    },
    'svc': {
        'model':SVC(),
        'params':{
        'kernel':['linear','rbf'],
        'C':[1,10,20]
          }
    }
}
```

```python
scores=[]
for model_name, mp in model_param.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_param': clf.best_params_
    })
```

```python
scores
```

```python
df=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
