```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('carprices.csv')
df
```

```python
pp=pd.get_dummies(df['Car Model'])
pp
```

```python
merged=pd.concat([df,pp],axis=1)
merged
merged=merged.drop(['Car Model','Mercedez Benz C class'],axis=1)
merged
```

```python
reg=LinearRegression()
reg.fit(merged[['Mileage','Age(yrs)','Audi A5','BMW X5']],merged['Sell Price($)'])
reg.predict(merged[['Mileage','Age(yrs)','Audi A5','BMW X5']])
reg.score(merged[['Mileage','Age(yrs)','Audi A5','BMW X5']],merged['Sell Price($)'])*100
```

```python
reg.predict([[69000,6,0,1]])
```
