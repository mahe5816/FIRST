```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import numpy as np
import matplotlib.pyplot as plt
```

```python
cancer=load_breast_cancer()
dir(cancer)
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df
```

```python
dft=cancer.target
dft
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
mm=sc.fit_transform(df)
scl=pd.DataFrame(mm,columns=df.columns)
scl
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scl,dft,test_size=0.3,random_state=105)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=21)
knn.fit(x_train,y_train)
p=knn.predict(x_test)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,p)
z=classification_report(y_test,p)
print(z)
```

```python
error= []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.score(x_test,y_test)
    error.append(pred_i)
```

```python
plt.plot(range(1,40),error,'--.')
plt.grid()
min(error)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
model_param= {
    'Logistic_Regression': {
        'model':LogisticRegression(),
        'params':{
        'C':[1,10,20]
          }
    },
    'svc': {
        'model':SVC(),
        'params':{
        'kernel':['linear','rbf'],
        'C':[1,10,20]
        }
    },
    'knn':{
        'model':KNeighborsClassifier(),
        'params':{
            'n_neighbors':[1,15,21]
           }
       }
}
```

```python
score=[]
for modelname,mp in model_param.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5)
    clf.fit(x_train,y_train)
    score.append({
        'model':modelname,
        'best_score':clf.best_score_,
        'param':clf.best_params_
    })
```
