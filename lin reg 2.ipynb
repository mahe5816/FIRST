```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([200,400,600,800,1200])
reg=LinearRegression()
reg.fit(x,y)
y_pred=reg.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred,c='red')
plt.show()
print("intercept",reg.intercept_)
print("coeff",reg.coef_)
print(reg.predict(np.array([6,7,8]).reshape(-1,1)))
import pickle
with open('mahe','wb') as f:
    pickle.dump(reg,f)
with open('mahe','rb') as f:
    mm=pickle.load(f)
mm.predict(np.array([1,2,3]).reshape(-1,1))
```
